# Importing Data

### Getting Rock Data

```python
RData = []
for filename in os.listdir(X_PATH+'rock/'):
    tempdata = np.loadtxt(X_PATH+'rock/'+filename,delimiter = ',')
    tempdata = tempdata[0:4,:13200]
    tempdata = np.reshape(tempdata,(4,60,220),order ='F')
    RData.append(tempdata)
    
RData = np.array(RData)
```

```python
RData.shape
```

### Getting Production Data

```python
PData = []
for filename in os.listdir(X_PATH+'production/'):
    Prod_unsized = np.loadtxt(X_PATH+'production/'+filename)
    Prod_resized = np.array([Prod_unsized[:,0]/3600.0/24.0, 
                             Prod_unsized[:,1]*3600*24,
                             Prod_unsized[:,2]*3600*24,
                             Prod_unsized[:,3]*3600*24,
                             Prod_unsized[:,4]*3600*24])
    PData.append(Prod_resized)
PData = np.array(PData)
```

```python
PData.shape
```

# Creating Dataset

```python
XRock = []
XProd = []
Y = []
WINDOWSIZE = 3

for index in range(RData.shape[0]):
    temprock = RData[index][0]
    temprock = np.expand_dims(temprock,axis = 2)
    for instance in range(PData.shape[2] - WINDOWSIZE):
        tempprod = PData[index,1:,instance:instance + WINDOWSIZE]
        tempprodNext = PData[index,1:,instance + WINDOWSIZE]
        tempprod = np.reshape(tempprod,(12,1))
#         print(tempprod,'\n')
#         print(tempprodNext,'\n')
#         if instance == 3:
#             print(tempprod.shape)
            
        XRock.append(temprock)
        XProd.append(tempprod)
        Y.append(tempprodNext)
```

```python
XRock = np.array(XRock)
XProd = np.array(XProd)
Y = np.array(Y)
```

## Splitting into Training and Testing

```python
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
XRock,XRocktest, XProd,XProdtest, Y, Ytest = train_test_split(XRock,XProd,Y,test_size=0.2,shuffle = False)

XRock, XProd, Y = shuffle(XRock, XProd, Y, random_state=0)
```

# Creating neural network architecture


```python
import numpy as np
import pandas as pd
import keras
import keras
from keras.layers import Reshape, Flatten, Input, Dense, concatenate, Conv2D, BatchNormalization, Activation, MaxPooling2D, Dropout, GlobalMaxPool2D
from keras.models import Model
from matplotlib import pyplot as plt
from scipy.ndimage import rotate as ro
```

```python
Input_figure = Input(shape=(60,220,1), name='input1')
Input_angle = Input(shape=(12,1), name = 'input2')

x = Conv2D(13, kernel_size=(3,3))(Input_figure)
x = BatchNormalization()(x)
x = Activation('elu')(x)
x = MaxPooling2D(pool_size=(2,2))(x)

x = Conv2D(10, kernel_size=(3,3))(x)
x = BatchNormalization()(x)
x = Activation('elu')(x)
x = MaxPooling2D(pool_size=(2,2))(x)

x = Conv2D(3, kernel_size=(3,3))(x)
x = BatchNormalization()(x)
x = Activation('elu')(x)
x = MaxPooling2D(pool_size=(2,2))(x)

x = Flatten()(x)
x = Reshape((375,1))(x)
#concatenate x and angle 
x = concatenate([x, Input_angle],axis = 1)
x = Flatten()(x)

x = Dense(30, activation='elu')(x)
out = Dense(4, activation='linear')(x)
```

```python
model = Model(inputs=[Input_figure, Input_angle], outputs=out)
```

```python
model.summary()
```

```python
opt = keras.optimizers.nadam()
model.compile(optimizer=opt,
              loss='mean_squared_error', 
            )
```

```python
history = model.fit([XRock,XProd],Y, batch_size = 50, epochs =5, verbose =1, validation_split = 0.1)
```

```python
model.save('TS3v1.h5')
```
