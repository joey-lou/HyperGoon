```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import random
import sklearn
import os
```

```python
X_PATH = 'C:/Users/Diyahir/Documents/SeniorDesign/caam-495-fluid-flow/New_Data/'
```

# Pressures

```python
data = np.loadtxt(X_PATH+'pressure/pressure1.txt',delimiter = ',')
data = np.reshape(data,(75,60,220),order ='F')
```

```python
plt.imshow(data[64,:,:])
plt.colorbar()
```

## Find mins and maxs

```python
mins = data.min(axis = 1).min(axis = 1)
maxs = data.max(axis = 1).max(axis = 1)
plt.plot(mins)
plt.plot(maxs)
plt.legend(['Mins','Max'])
plt.title('Pressure Mins and Max of Random Simulation')
plt.xlabel('Simulation itteration')
plt.ylabel('Value')
plt.show()
```

# Saturation

```python
satdata = np.loadtxt(X_PATH+'saturation/saturaton2.txt',delimiter = ',')
satdata = np.reshape(satdata,(75,60,220),order ='F')
```

```python
plt.imshow(satdata[74,:,:])
plt.colorbar()
```

```python
satmins = satdata.min(axis = 1).min(axis = 1)
satmaxs = satdata.max(axis = 1).max(axis = 1)
plt.plot(satmins)
plt.plot(satmaxs)
plt.legend(['Mins','Max'])
plt.title('Water saturation of Random Simulation')
plt.xlabel('Simulation itteration')
plt.ylabel('Value')
plt.show()
```
